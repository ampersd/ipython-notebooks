## Decision Tree

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
X = np.array([[1, 2], [3, 4], [5, 6]])
y = np.array([0, 1, 0])
clf = DecisionTreeClassifier()
clf.fit(X, y)
```

```python
importances = clf.feature_importances_
print importances
```

```python
import pandas
data = pandas.read_csv('titanic.csv', index_col='PassengerId')
res = data[['Pclass','Fare','Age','Sex','Survived']]
result = res[res.Age.notnull()]
result
```

```python
y = result['Survived']
y.head()
y.value_counts()
```

В данных есть пропущенные значения — например, для некоторых пассажиров неизвестен их возраст. Такие записи при чтении их в pandas принимают значение nan. Найдите все объекты, у которых есть пропущенные признаки, и удалите их из выборки.

```python
X = result[['Pclass','Fare','Age','Sex']]
X
```

```python
#male = 1, female = 0
d = {'male': 1, 'female': 0}
```

```python
#X = X[[X.Pclass,X.Fare,X.Age,X['Sex'].map(d)]]
X = X.replace(d)
X
```

```python
# help(DecisionTreeClassifier)
clf = DecisionTreeClassifier(random_state = 241)
clf.fit(X, y)
```

```python
importances = clf.feature_importances_
importances
```

**Ответ:**
Два признака, которые являеются наиболее значимыми для Survived (тех, кто выжил) - это
Fare (Цена билета) и Sex (Пол).
